```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
with open('experiment_perf/perf_git_single.data') as results:
    raw_trials = [re.search('m([0-9.]+)', result).group(1) for result in results.readlines() if result.startswith('real')]
```

```python
trials_single = pd.DataFrame(raw_trials, columns = ['time'])
trials_single['time'] = pd.to_numeric(trials_single['time'])
trials_single
```
